```python
from PIL import Image
import torchvision.transforms as transforms
import os
from fnmatch import fnmatch



# root = '/kaggle/input/plantaek6721/PlantaeK'
root = '/kaggle/input/plantdoc6721'
pattern = "*.jpg"

mean0 = 0
mean1 = 0
mean2 = 0

std0 = 0
std1 = 0
std2 = 0

ctr = 0


for path, subdirs, files in os.walk(root):
    for name in files:
        if fnmatch(name, pattern):
            img_path = os.path.join(path, name)
            
            print(img_path)

            # load the image
            img = Image.open(img_path)


            # define custom transform function
            transform_pre = transforms.Compose([
                transforms.ToTensor()
            ])
            img_tr = transform_pre(img)


            #############################################################

            # calculate mean and std
            mean, std = img_tr.mean([1,2]), img_tr.std([1,2])

            # print mean and std
#             print("mean and std before normalize:")
#             print("Mean of the image:", mean.tolist())
#             print("Std of the image:", std.tolist())

            mean0 += mean.tolist()[0]
            mean1 += mean.tolist()[1]
            mean2 += mean.tolist()[2]

            std0 += std.tolist()[0]
            std1 += std.tolist()[1]
            std2 += std.tolist()[2]

            ctr += 1
```

```python
new_mean = [round(mean0/ctr, 5), round(mean1/ctr, 5), round(mean2/ctr, 5)]
new_mean
```

```python
new_std = [round(std0/ctr, 5), round(std1/ctr, 5), round(std2/ctr, 5)]
new_std
```
